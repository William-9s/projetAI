# III- ENTRAINEMENT DU MODELE

```python
dataset=df.copy()
```

```python
# Sparer les variables indpendantes (X) de la variable dpendante (y)
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
# Diviser l'ensemble de donnes en ensembles d'entranement et de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
X_test.shape
```

#   quelques modeles 



 1) PREMIERE MODELE(K-NN)

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
```

```python
# Faire des prdictions sur l'ensemble de test
predictions = knn.predict(X_test)
```

```python

N,test_score,val_score=learning_curve(KNeighborsClassifier(n_neighbors=3),X_train,y_train,cv=5,train_sizes=np.linspace(0.1,1,10))
plt.plot(N,test_score.mean(axis=1),label="courbe de apprentissage")
plt.plot(N,val_score.mean(axis=1),label="courbe de test")
plt.title("courbe d'apprentissage et de test")
plt.legend(loc='lower right')
#plt.savefig("Images/knn_learning_curve.png");
```

#  IV-EVALUATION

```python
#score du modele1
print(classification_report(y_test,predictions))
```

```python
from sklearn import metrics
confusion_matrix=metrics.confusion_matrix(y_test,predictions)
cn_display=metrics.ConfusionMatrixDisplay(confusion_matrix=confusion_matrix)
cn_display.plot()
plt.show()
```

```python
liste=[]
for i in range(1,30):
    cls=KNeighborsClassifier(n_neighbors=i)
    cls.fit(X_train,y_train)
    y_pred=cls.predict(X_test)
    liste.append(1-accuracy_score(y_test,y_pred))
```

```python
plt.plot(range(1,30),liste)
plt.grid()
plt.title("Recherche du meilleur voisin")
plt.xlabel("valeur de k")
plt.ylabel("erreur");
```
